# ***LOADING SAMPLE***

```python
from google.colab import files
from PIL import Image

uploaded = files.upload()
image_path= list(uploaded.keys())[0]
image= Image.open(image_path)
```

# ***FRUIT ANALYSIS***

```python
def analyze_fruit(image):

  #CLASSIFY CATEGORY
  candidates_labels=["Fruit","Vegetable","Baked","Meat","Dairy"]
  labels= classifier(image , candidate_labels=candidates_labels)
  category=labels[0]["label"]
  if category != "Fruit":
    print("xxx Category Error xxx:\nOnly Fruits can be analysed in this classifier")
    return

  #VARIANCE CALCULATION
  import numpy as np
  img_array=np.array(image)
  img_array = img_array/255.0
  variance=np.var(img_array)

  #AVERAGE COLOR CALCULATION
  avg_color=np.mean(img_array, axis=(0,1))
  red, green , blue = avg_color

  #DETERMINE FRESHNESS
  if category == "Fruit":
      if variance > 0.08:
          Freshness = "Avoid"

      elif variance > 0.04:
          Freshness = "Okay"

      else:
          Freshness = "Fresh"

  return{
      "category": category,
      "variance": variance,
      "avg_color": avg_color,
      "Freshness": Freshness
  }

result= analyze_fruit(image)
print(result)
```
